# 3. Reduction operation

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
e
```

```python
# sum
torch.sum(e)
# sum along columns
torch.sum(e, dim=0)
# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
torch.mean(e)
# mean along col
torch.mean(e, dim=0)
```

```python
# median
torch.median(e)
```

```python
# max and min
torch.max(e)
torch.min(e)
```

```python
# product
torch.prod(e)
```

```python
# standard deviation
torch.std(e)
```

```python
# variance
torch.var(e)
```

```python
# argmax
torch.argmax(e)
```

```python
# argmin
torch.argmin(e)
```

#4. Matrix operations

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplcation
torch.matmul(f, g)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(f, 0, 1)
```

```python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
h
```

```python
# determinant
torch.det(h)
```

```python
# inverse
torch.inverse(h)
```

### 5. Comparison operations

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than equal to

# less than equal to
```
